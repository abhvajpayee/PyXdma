```python
import os
import sys
sys.path.insert(0, os.path.abspath('../../HDLRepo/src'))

from fhpc.tlpFormat import resetpkt

#pkt = resetpkt()
#dev.writeDMA(pkt, 0, 1)

def parsepkt(pkt):
    ptr = 0
    oldptr = -1
    pktfnd = 0
    res = []
    while ptr < len(pkt)-1:
        for i in range(ptr, len(pkt)):
            ptr = i
            if pkt[i] == 0xafafafafafafafaf:
                pktfnd = 1
                break
        if pktfnd:
            pkthead = ptr
            networkheader = int(pkt[pkthead + 1])
            Pktlen   = networkheader & 0xFFFF
            SlaveId  = (networkheader >> 16) & 0xFF
            MasterId = (networkheader >> 24) & 0xFF
            parserheader = int(pkt[pkthead + 2])
            IDENTIFIER = parserheader & 0xFF 
            enc = (parserheader >> 8) & 0xF
            nk  = (parserheader >> 12) & 0xF
            mode = (parserheader >> 16) & 0xFF
            size = (parserheader >> 24) & 0xFF
            packetid = (parserheader >> 48) & 0xFFFF
            res.append({'data':pkt[pkthead + 3: pkthead + 1 + Pktlen/8], 'packetid': packetid})
            ptr = pkthead + 1 + Pktlen/8
            pktfnd = 0
    return res
```
